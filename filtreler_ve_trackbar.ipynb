Filtreler
matplotlib kütüphanesi ile birden fazla grafikler gösterilebilir.
Resimlerin daha iyi işlenebilmesi için bazı filtreleme işlemleri yapılır.
subplot(121)-->bu komut 1 satır 2 sütundan oluşan bir plot hazırlar ve birinciyi çizdirir.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("opencv_logo.png")
kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title("2D Filtrelenmiş")
plt.xticks([]),plt.yticks([])
plt.show()
```

Blur işlemi de 2D benzer şekilde bulanıklaştırma işlemidir.
Bunun da çeşitleri vardır. Gaussian, median, bilateral vs

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("opencv_logo.png")
blur=cv2.GaussianBlur(img,(11,11),0)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title("Blurlanmış")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("opencv_logo.png")
median=cv2.medianBlur(img,9)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title("Blurlanmış")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("opencv_logo.png")
blur=cv2.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title("Blurlanmış")
plt.xticks([]),plt.yticks([])
plt.show()
```

MORFOLOJİK İŞLEMLER
Erosion, dilation, opening ve closing gibi morfolojik işlemler vardır.
Erozyon-->şeklin kenarlarını alır,
Dilate-->Şekli genişletir.

```python
import cv2
import numpy as np

img = cv2.imread('morf.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.imshow(img,cmap="gray")
plt.subplot(122),plt.imshow(erosion),plt.title("Erozyon")
plt.xticks([]),plt.yticks([])
plt.imshow(erosion,cmap="gray")
```

```python
import cv2
import numpy as np

img = cv2.imread('morf.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.imshow(img,cmap="gray")
plt.subplot(122),plt.imshow(dilation),plt.title("Dilate")
plt.xticks([]),plt.yticks([])
plt.imshow(dilation,cmap="gray")
```

```python
import cv2
import numpy as np

img = cv2.imread('open.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.imshow(img,cmap="gray")
plt.subplot(122),plt.imshow(opening),plt.title("Açma")
plt.xticks([]),plt.yticks([])
plt.imshow(opening,cmap="gray")
```

```python
import cv2
import numpy as np

img = cv2.imread('closing.png',0)
kernel = np.ones((7,7),np.uint8)
closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.xticks([]),plt.yticks([])
plt.imshow(img,cmap="gray")
plt.subplot(122),plt.imshow(dilation),plt.title("Kapama")
plt.xticks([]),plt.yticks([])
plt.imshow(closing,cmap="gray")
```

```python
import cv2
import numpy as np
def nothing(x):
    pass
img=np.zeros((300,650,3),np.uint8)
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
pencere_adi="HSV RENK PALET"
cv2.namedWindow(pencere_adi,cv2.WINDOW_AUTOSIZE)
# Track bar oluştur
cv2.createTrackbar("H",pencere_adi,0,179,nothing)
cv2.createTrackbar("S",pencere_adi,0,255,nothing)
cv2.createTrackbar("V",pencere_adi,0,255,nothing)
# döngüye gir
while(True):
    cv2.imshow(pencere_adi,img)
    tus=cv2.waitKey(1) & 0xFF
    if tus==ord("q"):
        break
    # track barların pozisyonunu al
    h=cv2.getTrackbarPos("H",pencere_adi)
    s=cv2.getTrackbarPos("S",pencere_adi)
    v=cv2.getTrackbarPos("V",pencere_adi)
    img[:]=[h,s,v]
    img=cv2.cvtColor(img,cv2.COLOR_HSV2BGR)
cv2.destroyAllWindows()
    
```
